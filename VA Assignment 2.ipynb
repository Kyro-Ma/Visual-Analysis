## Load Data and divide data into 12 slices based on sheets

```python
import numpy as np
import pandas as pd
import plotly.express as px 
import matplotlib.pyplot as plt


xls = pd.ExcelFile("/Users/kyroma/Downloads/Assignment 2 GHGHighlights Data.XLS")
sheetName = {'GHG Energy':'GHG emissions from energy', 
             'GHG Fugi':'GHG Fugitive emissions', 
             'GHG FC':'GHG emissions from fuel combustion', 
             'GHG FC - Coal':'GHG emissions from fuel combustion - Coal', 
             'GHG FC - Oil':'GHG emissions from fuel combustion - Oil', 
             'GHG FC - Gas':'GHG emissions from fuel combustion - Gas', 
             'CO2 MARBUNK':'CO2 emissions from international marine bunkers', 
             'CO2 AVBUNK':'CO2 emissions from international aviation bunkers', 
             'CO2-TES':'CO2 emissions / TES', 
             'CO2-GDP':'CO2 emissions / GDP using exchange rates', 
             'CO2-GDP PPP':'CO2 emissions / GDP using purchasing power parities', 
             'CO2-POP':'CO2 emissions / population'}

dictSheet = {} # extract 12 sheets

for name, value in sheetName.items():
    sheet = pd.read_excel(xls, name)
    dictSheet[name] = sheet
    
dictSheet
```

## Select Canada as country, then sort Canada data of every sheet from Year_2: maximum to Year_1: minimum

```python
regionData = {}

for name, value in sheetName.items():
    rows = dictSheet[name].loc[(dictSheet[name][value] == 'Region/Country/Economy') | (dictSheet[name][value] == 'Canada')]
    del rows[value]
    rowsSorted = rows.sort_values(by = rows.index[1], ascending=False, axis = 1) 
    rowsRenamed = rowsSorted.rename(index={rowsSorted.index[0]: "Year", rowsSorted.index[1]: "Data"})
    rowsRenamed = rowsRenamed.T
    rowsRenamed = rowsRenamed.reset_index(drop=True)
    rowsRenamed['Sheet'] = name
    regionData[name] = rowsRenamed

regionData
```

## Connect all dataframes into one, export to excel

```python
df = pd.DataFrame()

for name, value in regionData.items():
    df = pd.concat([df, regionData[name]])

df = df.reset_index(drop=True)
print(df)
df.to_excel('output.xlsx') 
```
